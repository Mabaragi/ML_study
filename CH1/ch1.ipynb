```python
# 방어 데이터

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

plt.scatter(x=smelt_length, y=smelt_weight)
```

```python
fish_df = pd.DataFrame(
    {
        'length' : bream_length + smelt_length,
        'weight' : bream_weight + smelt_weight
    }
)
```

### 정답 리스트
K-최근접 이웃 알고리즘을 사용하기 위해서, 무엇이 정답인지 알려줄 데이터가 필요하다.

도미는 1이고 방어는 0으로 표현한다.

```python
fish_df['type'] = [1] * 35 + [0] * 14
fish_df.to_csv('./fish_data.csv')
fish_df
```

```python
from sklearn.neighbors import KNeighborsClassifier
# 모델 객체 생성
kn = KNeighborsClassifier()
```

타겟과 특성을 이용하여 훈련을 시킴. 사이킷런에서는 `fit()` 메서드로 할 수 있다.

```python
# kn 모델로 알고리즘 훈련
kn.fit(fish_df[['weight', 'length']], fish_df['type'])
```

```python
#  모델이 잘 훈련되었는지 평가. 0~에서 1 사이의 값을 가짐.
kn.score(fish_df[['weight', 'length']], fish_df['type'])
```

정확도가 1.0이면 도미와 방어를 완벽하게 분류한 것.

### K-최근접 이웃 알고리즘
어떤 데이터에 대한 답을 구할 떄, 주위의 다른 데이터를 보고 다수를 차지하는 것을 정답으로 사용하는 알고리즘.

```python
# 무게 600, 길이 30인 생선의 type을 예측.
kn.predict([[600, 30]])
```

K-최근접 이웃 알고리즘은 데이터가 많은 경우에 사용하기 힘들다.

K-최근접 이웃 알고리즘은 기본적으로 가장 가까운 5개의 데이터를 참고로한다. 매개변수를 통해 이것을 바꿀 수 있다.

```python
# 49개로 설정하면, 대부분인 35개가 도미이므로 전부 도미로 판단한다.
kn_49 = KNeighborsClassifier(n_neighbors=49)
kn_49.fit(fish_df[['length', 'weight']], fish_df['type'])
kn_49.score(fish_df[['length', 'weight']], fish_df['type'])
```

```python
# 따라서 35/49의 확률로 정답을 예측하며, 모든 데이터의 타입을 1로 예측한다.
kn_49.predict([[1, 1]])
```
